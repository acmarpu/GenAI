```python
#2
# # Tokenization with TikToken
# The cat sat on the mat  
# python -m pip install tiktoken

import tiktoken

encoder = tiktoken.encoding_for_model("gpt-4o")

text = "The cat sat on the mat"

tokens = encoder.encode(text)

print("Text:", text)

print("length of tokens", len(tokens))

for token_id in tokens:
    print(f"Token ID: {token_id}, Token: {encoder.decode([token_id])}")
```

```python
#6
# # Setup Google GenAI Client
import os
from dotenv import load_dotenv
from google import genai

load_dotenv()
os.environ["GEMINI_API_KEY"] = os.getenv("GEMINI_API_KEY")
client = genai.Client(api_key=os.environ["GEMINI_API_KEY"])
print("Client Ready!")

# Chatbot

def chat():
    history = []

    # system instructions
    system_instruction = "You are a friendly chatbot."

    # it should keep asking questions and talk to us
    while True:
        try:
            user_input = input("you: ")
        except (KeyboardInterrupt, EOFError):
            print("\nGoodbye!")
            break
    
    # if user entered by or quit - we need to exit the chatbot

        if user_input.strip() == "quit" or user_input.strip() == "bye":
            print("Goodbye!")
            break

    # remember what user asked
        history.append({"role": "user", "parts": [
            {"text": user_input}
        ]})


        response = client.models.generate_content(
            model="gemini-3.5-flash",
            contents=history,
            config={
                "system_instruction": system_instruction
            }
        )

        assistant_response = response.text

        
        history.append({"role": "model", "parts": [
            {"text": assistant_response}
        ]})
         

        print(f"bot: {assistant_response}")
```
